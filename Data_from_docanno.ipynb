```python
data_previous_annotates=pd.read_csv('../../whatsapp-analytics/all_80_annotated.csv')
data_previous_annotates
```

```python
total_data_annotated['Punyajoy']=data_previous_annotates['Punyajoy']
total_data_annotated['Mithun']=data_previous_annotates['Mithun']
total_data_annotated['Soumyadeep']=data_previous_annotates['Soumyadeep']
total_data_annotated
```

```python
from tqdm import tqdm

map_dict={'Fear speech':1,'Normal':0}

user_dict={1:'Punyajoy',36:'Mithun',37:'Soumyadeep'}

list_of_rows=[]
list_of_rows_for_test=[]
from collections import Counter 
  
def most_frequent(List): 
    occurence_count = Counter(List) 
    try:
        #print(occurence_count.most_common(1)[0][0],occurence_count.most_common(1)[0][1])
        return occurence_count.most_common(1)[0][0],occurence_count.most_common(1)[0][1] 
    except  IndexError:
        return 0,0


list_not_found=[]

for i in tqdm(range(len(text_data))):
    found_id=re.search(r'\$\#(.*?)\$\#', text_data[i]['text']).group(1)
    annotation=text_data[i]['annotations']
    dict_row={'id':found_id}
    
    label_list=[]
    for anno in annotation:
        if(dict_label[anno['label']] in ['Fear speech','Normal']):
            dict_row[user_dict[anno['user']]]=map_dict[dict_label[anno['label']]]
            label_list.append(map_dict[dict_label[anno['label']]])
    
    label,count=most_frequent(label_list)
    if(count>1 and len(label_list)==3):
        print(count,label)
        dict_row['label']=label
        dict_row['text']=total_data.loc[int(found_id),:]['text']
        dict_row['translated']=total_data.loc[int(found_id),:]['translated']
        list_of_rows.append(dict_row)
    elif (len(label_list)>1):
        print("id not completed",found_id)
        list_not_found.append(found_id)
```

```python
labeled_data=pd.DataFrame(list_of_rows)
```
